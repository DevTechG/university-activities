```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow 
```

```python
#Espelhamento
image = np.ones((400, 400, 3), np.uint8) * 255
point1 = (150, 50)
point2 = (50, 150)
point3 = (150, 150)

cv.line(image, point1, point2, (0, 255, 0), 6)
cv.line(image, point2, point3, (0, 255, 0), 6)
cv.line(image, point3, point1, (0, 255, 0), 6)

res = np.ones((400, 400, 3), np.uint8) * 255
cv.line(res, (int(point1[0]+(point1[1])), int(point1[1])), (int(point2[0]+(point2[1])), int(point2[1])), (0, 255, 0), 6)
cv.line(res, (int(point2[0]+(point2[1])), int(point2[1])), (int(point3[0]+(point3[1])), int(point3[1])), (0, 255, 0), 6)
cv.line(res, (int(point3[0]+(point3[1])), int(point3[1])), (int(point1[0]+(point1[1])), int(point1[1])), (0, 255, 0), 6)
res2=cv.flip(image,1)

cv2_imshow(image)
cv2_imshow(res)
cv2_imshow(res2)
```

```python
#cisalhamento
image = np.ones((400, 400, 3), np.uint8) * 255
point1 = (150, 50)
point2 = (50, 150)
point3 = (150, 150)

cv.line(image, point1, point2, (0, 255, 0), 6)
cv.line(image, point2, point3, (0, 255, 0), 6)
cv.line(image, point3, point1, (0, 255, 0), 6)
sx = 1.5

res = np.ones((400, 400, 3), np.uint8) * 255
cv.line(res, (int(point1[0]+(point1[1]*sx)), int(point1[1])), (int(point2[0]+(point2[1]*sx)), int(point2[1])), (0, 255, 0), 6)
cv.line(res, (int(point2[0]+(point2[1]*sx)), int(point2[1])), (int(point3[0]+(point3[1]*sx)), int(point3[1])), (0, 255, 0), 6)
cv.line(res, (int(point3[0]+(point3[1]*sx)), int(point3[1])), (int(point1[0]+(point1[1]*sx)), int(point1[1])), (0, 255, 0), 6)

cv2_imshow(image)
cv2_imshow(res)
```
