```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
#Criando os vetores x e y
x = np.array(range(100))
y = x ** 2

#Criando o plot
plt.plot([1, 2, 3, 4])
plt.plot(x, (1/2) * y + 7, label = 'y = 0.5 * x^2 + 7')
plt.plot(x, y + 1000, label = 'y = x^2 + 1000')

plt.title('Meu primeiro gráfico')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(alpha=0.4, linestyle='--')
#{'-', '--', '-.', ':', '',(offset, on-off-seq), ...}
plt.plot(x,y,label='y = x^2')
plt.legend()

#Mostrando o plot
plt.show()

#Mostrando imagem com o plot
img = cv.imread('DER_CompanhiaAérea_PeterChen.png')
plt.axis('on')
plt.imshow(img)

```
