```python
def linha(x1: int, y1: int, x2: int, y2: int) -> None:
  m = (y2 - y1) / (x2 - x1)
  b = y1 - m * x1
  print('X = %.2f Y = %.2f' %(x1,y1))
  while (x1 < x2):
    x = x1 + 1
    y = m * x + b
    x1 = x
    print('X = %.2f Y = %.2f' %(x,y))

linha(6,9,11,12)
```
