```python
#2. Escreva um programa que receba como parâmetros o centro, o raio e uma cor e desenhe o círculo usando o algoritmo simples de simetria de octantes. 

def midPoint(x, r):
    x = 0
    y = r
    vetX= [0]
    vetY= [r]
    
    p = 1 - r
    
    while x < y:
        x += 1
        if p <= 0:
            p = p + 2 * x + 1
        else:        
            y -= 1
            p = p + 2 * x - 2 * y + 1
        #Oct1
        vetX.append(x)
        vetY.append(y)

    tam = len(vetX)

    #Oct8        
    for i in range(tam-1, -1, -1):
        vetX.append(vetX[i])
        vetY.append(-vetY[i])

    #Oct5        
    for i in range(0, tam):
        vetX.append(-vetX[i])
        vetY.append(-vetY[i])

    #Oct4        
    for i in range(tam-1, -1, -1):
        vetX.append(-vetX[i])
        vetY.append(vetY[i])

    #Oct2        
    for i in range(tam-1, -1, -1):
         vetX.append(vetY[i])
         vetY.append(vetX[i])

    #Oct3        
    for i in range(0, tam):
        vetX.append(vetY[i])
        vetY.append(-vetX[i])

    #Oct7        
    for i in range(tam-1, -1, -1):
        vetX.append(-vetY[i])
        vetY.append(vetX[i])

    #Oct6        
    for i in range(0, tam):
        vetX.append(-vetY[i])
        vetY.append(-vetX[i])
    
    return vetX, vetY


def plotCircle (x,y,r,g,b):
    plt.plot(x,y, color = (r,g,b))
    plt.show()


x, y = midPoint(0,10)

print(x)
print(y)

plotCircle(x,y,0.5,0.5,1)

plt.show(plt.plot(x,y))
```
