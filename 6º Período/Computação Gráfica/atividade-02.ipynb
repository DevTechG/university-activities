Exercício 1

Defina representação analítica de curvas.
```
```
A representação analítica de curvas utiliza de uma ou várias equações para apresentar uma curva, tendo suas vantagens em comparação com uma representação por conjuntos de pontos: ser mais precisa, não requerer área de armazenamento, facilidade de calcular novos pontos, maior facilidade de obter propriedades da curva como inclinação e área. Podendo ser paramétrica ou não-paramétrica. A não-paramétrica não recebe parâmetros, sendo calculado através de uma equação de y em x ou vice-versa. Todos os pontos adicionais a serem utilizados para formar as retas são valores exatos e não aproximações dos valores. Esse método de representação, também apresenta vantagens quando se trata de desenhar transformações como mudanças de escala, rotação, projeções e outras transformadas.

Exercício 2

Diferencie representações analíticas paramétricas e não-paramétricas. 
```
```
Na forma não-paramétrica, como o nome indica, não há parâmetros, de modo que y é dado como uma função de x e vice-versa, de modo que y = fx(x) e x  = fy(y). 
Neste modelo de representação, uma equação de reta fica: 
y = 2 * x - 1 ou x = ½ * (y + 1). 
A forma não-paramétrica de representar uma curva é dada por uma equação do tipo y = f(x), onde os pares (x,y = f(x)) representam os pontos da curva, de forma que uma equação explícita de uma parábola é dada por:
y = a * x² + b * x + c, 
sendo a, b e c números quaisquer.
Na forma paramétrica, usa-se um parâmetro (t, θ, etc.) para definir as coordenadas dos pontos da curva. Neste modelo de representação uma reta pode ser representada por: 
x = t + 1 = fx(t) e y = 2 * t + 1 = fy(t).
Os parâmetros usados nessa representação podem ser qualquer símbolo. Nesta forma cada coordenada de um ponto em uma curva é representada como uma função de um único parâmetro, sendo que a posição na curva é fixada pelo valor do parâmetro. Uma curva 2D que usa t como parâmetro, as coordenadas são funções desse t: 
x = x(t) e y = y(t), 
com a posição do ponto sendo dada por P(t) = (x(t), y(t)). Desse modo a maior diferença entre as duas formas de representação é que na paramétrica utiliza-se de algum valor como função  representando uma parábola com 
**x = a * t² e y = 2 * a * t** 
e a forma não-paramétrica, não possui esse valor de t para calcular a posição do ponto e portanto uma parábola seria dada por 
**y² - 4 * a * x = 0**

````python
#Exercício 7
#Projete um círculo usando a equação: x²+ y²−r=0 Onde r é um raio qualquer.

x = np.linspace(-10 , 10, 180) 
y = np.linspace(-10 , 10, 180) 
r = 50
a, b = np.meshgrid(x , y) 
  
C = a ** 2 + b ** 2 - r
  
figure, axes = plt.subplots() 
  
axes.contour(a, b, C , [0]) 
axes.set_aspect(1) 
  
plt.show()
````

````python
#Exercício 8
#Projete uma reta usando a seguinte equação: 2x−y−1=0

x = np.linspace(0 , 10, 100)
y = np.linspace(0 , 10, 100)
a, b = np.meshgrid(x , y)
  
Fx = 2 * a - b - 1
  
figure, axes = plt.subplots()
  
axes.contour(a, b, Fx , [0])
axes.set_aspect(1)
  
plt.show()
````
