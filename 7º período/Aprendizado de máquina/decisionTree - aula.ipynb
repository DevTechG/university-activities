```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

df = pandas.read_csv("data.csv")

# print(df)
df
```

```python
mapNationality = {'UK': 0, 'USA': 1, 'N': 2}

df['Nationality'] = df['Nationality'].map(mapNationality)

# print(df['Nationality'])
```

```python
mapGo = {'YES': 1, 'NO': 0}

df['Go'] = df['Go'].map(mapGo)

# print(df)
```

```python
features =  ['Age', 'Experience', 'Rank', 'Nationality']

x = df[features]
y = df['Go']
print(x)
print(y)
```

```python
dTree = DecisionTreeClassifier()

dTree = dTree.fit(x, y)
tree.plot_tree(dTree, feature_names = features)
```

```python
print(dTree.predict([[0, 0, 6.6, 0]]))
```
