```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import linear_model
from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

#Armazenando os atributos na matriz X
xIris = iris.data

#Armazenando o vetor alvo em Y
yIris = iris.target
```

```python
sepalLength = xIris[ :, 0]
sepalWidth  = xIris[ :, 1]
petalLength = xIris[ :, 2]
petalWidth  = xIris[ :, 3]
```

```python
plt.scatter(sepalLength, yIris)
plt.title('Comprimento Sépala - Sepal length')
plt.show()

plt.scatter(sepalWidth, yIris)
plt.title('Largura Sépala - Sepal width')
plt.show()

plt.scatter(sepalLength, yIris)
plt.title('Comprimento Pétala - Petal length')
plt.show()

plt.scatter(sepalWidth, yIris)
plt.title('Largura Pétala - Petal width')
plt.show()
```

```python
modelSepalLength = numpy.poly1d(numpy.polyfit(sepalLength, yIris, 3))
modelSepalWidth = numpy.poly1d(numpy.polyfit(sepalWidth, yIris, 3))
modelPetalLength = numpy.poly1d(numpy.polyfit(petalLength, yIris, 3))
modelPetalWidth = numpy.poly1d(numpy.polyfit(petalWidth, yIris, 3))

myline = numpy.linspace(1, 5, 100)

plt.scatter(sepalWidth, yIris)
plt.plot(myline, modelSepalLength(myline))
plt.title('Regressão polinomial - comprimento sépala')
plt.show()

plt.scatter(sepalWidth, yIris)
plt.plot(myline, modelSepalWidth(myline))
plt.title('Regressão polinomial - largura sépala')
plt.show()

plt.scatter(sepalWidth, yIris)
plt.plot(myline, modelPetalLength(myline))
plt.title('Regressão polinomial - comprimento pétala')
plt.show()

plt.scatter(sepalWidth, yIris)
plt.plot(myline, modelPetalWidth(myline))
plt.title('Regressão polinomial - largura pétala')
plt.show()
```

```python
print('R-quadrado', r2_score(yIris, modelSepalLength(sepalWidth)))
print('R-quadrado', r2_score(yIris, modelSepalWidth(sepalWidth)))
print('R-quadrado', r2_score(yIris, modelPetalLength(petalLength)))
print('R-quadrado', r2_score(yIris, modelPetalWidth(petalWidth)))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
iris_X, iris_y = load_iris(return_X_y=True)

# Use only one feature
iris_X = iris_X[:, np.newaxis, 2]

# Split the data into training/testing sets
iris_X_train = iris_X[:-20]
iris_X_test = iris_X[-20:]

# Split the targets into training/testing sets
iris_y_train = iris_y[:-20]
iris_y_test = iris_y[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(iris_X_train, iris_y_train)

# Make predictions using the testing set
iris_y_pred = regr.predict(iris_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(iris_y_test, iris_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(iris_y_test, iris_y_pred))

# Plot outputs
plt.scatter(iris_X_test, iris_y_test, color="red")
plt.plot(iris_X_test, iris_y_pred, color="green", linewidth=3)

plt.scatter(iris_X_test, iris_y_test)
plt.plot(iris_X_test, iris_y_pred)


plt.xticks(())
plt.yticks(())

plt.show()
```
