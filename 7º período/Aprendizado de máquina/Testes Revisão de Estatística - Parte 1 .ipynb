(Enem 2015) Uma pessoa, ao fazer uma pesquisa com alguns alunos de um curso, coletou as idades dos entrevistados e organizou esses dados em um gráfico.

image.png

Qual a moda das idades, em anos, dos entrevistados?

```python
from scipy import stats

ages = []

for x in range(21):
  ages.append(9)

for x in range(15):
  ages.append(12)

for x in range(12):
  ages.append(18)

modeAges = stats.mode(ages)

print("A moda das idadesé", modeAges)
```

Os valores dos salários dos funcionários de uma empresa estão representados na tabela a seguir:

Analisando a tabela de salários dos funcionários da empresa, podemos afirmar que:

A) a moda salarial dessa empresa é R$ 3.525,00, que é o salário dos consultores.

B) a mediana dos salários é de R$ 6.391,40.

C) a maioria dos funcionários ganha abaixo de R$ 3.525,00.

D) a moda salarial dessa empresa é o conjunto {2, 4}.

E) a média salarial dessa empresa é R$ 3.525,00.

```python
import numpy
from scipy import stats

wage = [44500.00, 1212]

for x in range(2):
  wage.append(18650)

for x in range(4):
  wage.append(9257.80)

for x in range(80):
  wage.append(3525)

for x in range(2):
  wage.append(1980.27)

modeWage = stats.mode(wage)
medianWage = numpy.median(wage)
meanWage = numpy.mean(wage)

print("A moda dos salários informados é:", modeWage, "\nA mediana dos salários informados é:", medianWage,"\nA média dos salários informados é:", meanWage)
```

Três alunos, X, Y e Z, estão matriculados em um curso de inglês. Para avaliar esses alunos, o professor optou por fazer cinco provas. Para que seja aprovado nesse curso, o aluno deverá ter a média aritmética das notas das cinco provas maior ou igual a 6. Na tabela, estão dispostas as notas que cada aluno tirou em cada prova.

image.png

Com base nos dados da tabela e nas informações dadas, ficará(ão) reprovado(s)

```python
import numpy

averageGrade = []

studentGradeX= [5, 5, 5, 10, 6]
studentGradeY= [4, 9, 3, 9, 5]
studentGradeZ= [5, 5, 8, 5, 6]

averageGrade.append(numpy.mean(studentGradeX))
averageGrade.append(numpy.mean(studentGradeY))
averageGrade.append(numpy.mean(studentGradeZ))

for x in range (3):
    if averageGrade[x] < 6:
        print("O aluno", x, " reprovou com média", averageGrade[x],"\n")
```

Os candidatos K, L, M, N e P estão disputando uma única vaga de emprego em um empresa e fizeram provas de português, matemática, direito e informática. A tabela apresenta as notas obtidas pelos cinco candidatos.

Candidato K - 0
Candidato L - 1
Candidato M - 2
Candidato N - 3
Candidato P - 4


image.png

Segundo o edital de seleção, o candidato aprovado será aquele para o qual a mediana das notas obtidas por ele nas quatro disciplinas for a maior. O candidato aprovado será:

```python
import numpy

result = [[33, 33, 33, 34], [32, 39, 33, 34], [35, 35, 36, 34], [24, 37, 40, 35], [36, 16, 26, 41]]
medianResult = []
candidateApproved = 0

for x in range (5):
    medianResult.append(numpy.median(result[x]))

for x in range (5):
    if  medianResult[candidateApproved] < medianResult[x]:
        candidateApproved = x

print("O candido aprovado é o ", candidateApproved, "com mediana", medianResult[candidateApproved] )
```
