# **Regressão Polinomial**

Se seus pontos de dados claramente não se ajustarem a uma regressão linear (uma linha reta através de todos os pontos de dados), pode ser ideal para regressão polinomial.

A regressão polinomial, como a regressão linear, usa a relação entre as variáveis ​​x e y para encontrar a melhor maneira de traçar uma linha através dos pontos de dados.

img_polynomial_regression.png

# **Como Funciona?**

Python tem métodos para encontrar uma relação entre pontos de dados e para desenhar uma linha de regressão polinomial. Mostraremos como usar esses métodos em vez de passar pela fórmula matemática.

No exemplo abaixo, cadastramos 18 carros passando por um determinado posto de pedágio.

Registramos a velocidade do carro e a hora do dia (hora) em que ocorreu a passagem.

O eixo x representa as horas do dia e o eixo y representa a velocidade:

```python
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22] # hora do dia
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100] # velocidade do carro

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

# **R-quadrado**

É importante saber quão bem é a relação entre os valores dos eixos x e y, se não houver relação, a regressão polinomial não pode ser usada para prever nada.

A relação é medida com um valor chamado r-quadrado.

O valor de r ao quadrado varia de 0 a 1, onde 0 significa nenhuma relação e 1 significa 100% relacionado.

O Python e o módulo Sklearn calcularão esse valor para você, tudo o que você precisa fazer é alimentá-lo com os arrays x e y.

Quão bem meus dados se encaixam em uma regressão polinomial?

```python
import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```


**Nota:** O resultado 0,94 mostra que existe uma relação muito boa e podemos usar a regressão polinomial em previsões futuras.

# **Prever Valores Futuros**

Agora podemos usar as informações que reunimos para prever valores futuros.

Exemplo: Vamos tentar prever a velocidade de um carro que passa no posto de pedágio por volta das 17:00:

Para fazer isso, precisamos do mesmo array mymodel do exemplo acima:

```python
import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

speed = mymodel(17)
print(speed)
```

O exemplo previu uma velocidade de 88,87, que também podemos ler no diagrama:

img_polynomial_prediction.png

Ajuste Ruim?

Vamos criar um exemplo onde a regressão polinomial não seria o melhor método para prever valores futuros.

Esses valores para os eixos x e y devem resultar em um ajuste muito ruim para a regressão polinomial:

```python
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(2, 95, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

print(r2_score(y, mymodel(x)))
```

O resultado: 0,00995 indica um relacionamento muito ruim e nos informa que esse conjunto de dados não é adequado para regressão polinomial.

Podemos prever a emissão de CO2 de um carro com base no tamanho do motor, mas com a regressão múltipla podemos incluir mais variáveis, como o peso do carro, para tornar a previsão mais precisa.

# **Como Funciona?**

Em Python, temos módulos que farão o trabalho para nós. 

1 - Comece importando o módulo Pandas.

2 - Leia o arquivo com os dados da tabela abaixo.

3 - Em seguida, faça uma lista dos valores independentes e chame essa variável de X. 

4 - Coloque os valores dependentes em uma variável chamada y.

5 - Importe o módulo sklearn.

6 - A partir do módulo sklearn, usaremos o método LinearRegression() para criar um objeto de regressão linear. Este objeto possui um método chamado fit() que toma os valores independentes e dependentes como parâmetros e preenche o objeto de regressão com dados que descrevem o relacionamento.

7 - Agora temos um objeto de regressão pronto para prever os valores de CO2 com base no peso e volume de um carro.

Veja o exemplo completo abaixo.


```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("https://www.w3schools.com/python/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[2300, 1300]])

print(predictedCO2)
```

Prevemos que um carro com motor de 1,3 litro e peso de 2.300 kg liberará aproximadamente 107 gramas de CO2 por quilômetro rodado.

# **Coeficiente**

O coeficiente é um fator que descreve a relação com uma variável desconhecida.

Exemplo: se x é uma variável, então 2x é x duas vezes. x é a variável desconhecida e o número 2 é o coeficiente.

Nesse caso, podemos pedir o valor do coeficiente de peso em relação ao CO2 e de volume em relação ao CO2. A(s) resposta(s) que obtemos nos diz o que aconteceria se aumentássemos ou diminuíssemos um dos valores independentes.

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("https://www.w3schools.com/python/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

print(regr.coef_)
```

# **Resultado Explicado**

A matriz de resultados representa os valores dos coeficientes de peso e volume.

**Peso:** 0,00755095 

**Volume:** 0,00780526

Estes valores dizem-nos que se o peso aumentar 1kg, a emissão de CO2 aumenta 0,00755095g.

E se o tamanho do motor (Volume) aumentar em 1 cm<sup>3</sup>, a emissão de CO2 aumentará em 0,00780526 g.

Acho que é um palpite justo, mas vamos testar!

Já previmos que se um carro com motor de 1300cm<sup>3</sup> pesar 2300kg, a emissão de CO2 será de aproximadamente 107g.

E se aumentarmos o peso em 1000kg?

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("https://www.w3schools.com/python/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

predictedCO2 = regr.predict([[3300, 1300]])

print(predictedCO2)
```

Prevemos que um carro com motor de 1,3 litro e 3.300 kg de peso liberará aproximadamente 115 gramas de CO2 por quilômetro rodado.

O que mostra que o coeficiente de 0,00755095 está correto:

107,2087328 + (1000 * 0,00755095) = 114,75968

Pode ser difícil comparar o volume 1.0 com o peso 790, mas se escalarmos ambos em valores comparáveis, podemos ver facilmente quanto um valor é comparado ao outro.

Existem diferentes métodos para dimensionar dados, neste tutorial usaremos um método chamado padronização.

O método de padronização usa esta fórmula:

z = (x - u) / s

Onde z é o novo valor, x é o valor original, u é a média e s é o desvio padrão.

Se você pegar a coluna de peso do conjunto de dados acima, o primeiro valor será 790 e o valor escalado será:

(790 - 1292.23) / 238.74 = -2.1

Se você pegar a coluna de volume do conjunto de dados acima, o primeiro valor será 1,0 e o valor escalado será:

(1.0 - 1.61) / 0.38 = -1.59

Agora você pode comparar -2,1 com -1,59 em vez de comparar 790 com 1,0.

Você não precisa fazer isso manualmente, o módulo Python sklearn possui um método chamado StandardScaler() que retorna um objeto Scaler com métodos para transformar conjuntos de dados.

```python
import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

df = pandas.read_csv("https://www.w3schools.com/python/data.csv")

X = df[['Weight', 'Volume']]

scaledX = scale.fit_transform(X)

print(scaledX)
```

A tarefa no capítulo de Regressão Múltipla era prever a emissão de CO2 de um carro quando você conhecia apenas seu peso e volume.

Quando o conjunto de dados for dimensionado, você terá que usar a escala ao prever valores:



```python
import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

df = pandas.read_csv("https://www.w3schools.com/python/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

scaledX = scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaledX, y)

scaled = scale.transform([[2300, 1.3]])

predictedCO2 = regr.predict([scaled[0]])
print(predictedCO2)
```

# **Avalie Seu Modelo**

Em Machine Learning criamos modelos para prever o resultado de certos eventos, como no capítulo anterior onde previmos a emissão de CO2 de um carro quando sabíamos o peso e a cilindrada do motor.

Para medir se o modelo é bom o suficiente, podemos usar um método chamado Train/Test.

**O que é Treinar/Testar?**

Train/Test é um método para medir a precisão do seu modelo.

Chama-se Train/Test porque você divide o conjunto de dados em dois conjuntos: um conjunto de treinamento e um conjunto de teste.

80% para treinamento e 20% para teste.

Você treina o modelo usando o conjunto de treinamento.

Você testa o modelo usando o conjunto de teste.

Treinar o modelo significa criar o modelo.

Testar o modelo significa testar a precisão do modelo.

## **Comece com um conjunto de dados**

Comece com um conjunto de dados que você deseja testar.

Nosso conjunto de dados ilustra 100 clientes em uma loja e seus hábitos de compra.


```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

# **Ajustar o Conjunto de Dados**

Como é o conjunto de dados? Na minha opinião, acho que o melhor ajuste seria uma regressão polinomial, então vamos desenhar uma linha de regressão polinomial.

Para desenhar uma linha através dos pontos de dados, usamos o método plot() do módulo matplotlib:

```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

O resultado pode apoiar minha sugestão de que o conjunto de dados se ajusta a uma regressão polinomial, embora nos dê alguns resultados estranhos se tentarmos prever valores fora do conjunto de dados. Exemplo: a linha indica que um cliente que passasse 6 minutos na loja faria uma compra no valor de 200. Isso provavelmente é um sinal de overfitting.

Mas e a pontuação do R ao quadrado? A pontuação do R-quadrado é um bom indicador de quão bem meu conjunto de dados está se ajustando ao modelo.

# **R2**

Lembre-se de R2, também conhecido como R-quadrado?

Ele mede a relação entre o eixo x e o eixo y, e o valor varia de 0 a 1, onde 0 significa nenhuma relação e 1 significa totalmente relacionado.

O módulo sklearn possui um método chamado r2_score() que nos ajudará a encontrar essa relação.

Neste caso, gostaríamos de medir a relação entre os minutos que um cliente permanece na loja e quanto dinheiro ele gasta.

```python
import numpy
from sklearn.metrics import r2_score
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```

**Nota:** O resultado 0,799 mostra que existe uma relação OK.

# **Traga o Conjunto de Teste**

Agora criamos um modelo que está OK, pelo menos quando se trata de dados de treinamento.

Agora queremos testar o modelo com os dados de teste também, para ver se nos dá o mesmo resultado.

```python
import numpy
from sklearn.metrics import r2_score
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

r2 = r2_score(test_y, mymodel(test_x))

print(r2)
```

**Observação:** o resultado 0,809 mostra que o modelo também se ajusta ao conjunto de teste e estamos confiantes de que podemos usar o modelo para prever valores futuros.
