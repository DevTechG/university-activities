```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris= load_iris()

X= iris.data
y= iris.target

y = y.reshape(-1, 1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X[:,0:2], y, test_size = 0.20) # Divisão dos dados.

degree = 2

model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X_train, y_train)
predictions = model.predict(X_test)

# Plotando o conjunto de treinamento de acordo com as classes.

ind = np.where(y_train == 0)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind,0], X_train[ind,1], color='navy', s=30, marker='o')

ind = np.where(y_train == 1)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind,0], X_train[ind,1], color='turquoise', s=30, marker='o')

ind = np.where(y_train == 2)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind,0], X_train[ind,1], color='indigo', s=30, marker='o')

# Plotando o conjunto de teste de acordo com as previsões feitas.

predictions = np.round(np.abs(predictions))
ind = np.where(predictions == 0.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind,0], X_test[ind,1], color='purple', s=30, marker='v')

ind = np.where(predictions == 1.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind,0], X_test[ind,1], color='yellow', s=30, marker='P')

ind = np.where(predictions == 2.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind,0], X_test[ind,1], color='salmon', s=30, marker='D')

# Plotando o conjunto de teste de acordo com o esperado.

ind = np.where(y_test == 0.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind,0], X_test[ind,1], color='violet', s=30, marker='^')

ind = np.where(y_test == 1.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind,0], X_test[ind,1], color='lime', s=30, marker='X')

ind = np.where(y_test == 2.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind,0], X_test[ind,1], color='brown', s=30, marker='d')

plt.show()

score = r2_score(y_test, predictions)
print(score)

rmse = np.sqrt(((predictions - y_test)**2).mean())
print(rmse)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

# Divisão dos dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X[:, 2:4], y, test_size=0.20)

degree = 2

# Criação do modelo e treinamento
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X_train, y_train)
predictions = model.predict(X_test)

# Plotando o conjunto de treinamento de acordo com as classes
ind = np.where(y_train == 0)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind, 0], X_train[ind, 1], color='navy', s=30, marker='o')

ind = np.where(y_train == 1)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind, 0], X_train[ind, 1], color='turquoise', s=30, marker='o')

ind = np.where(y_train == 2)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind, 0], X_train[ind, 1], color='indigo', s=30, marker='o')

# Plotando o conjunto de teste de acordo com as previsões feitas
predictions = np.round(np.abs(predictions))
ind = np.where(predictions == 0.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color='purple', s=30, marker='v')

ind = np.where(predictions == 1.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color='yellow', s=30, marker='P')

ind = np.where(predictions == 2.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color='salmon', s=30, marker='D')

# Plotando o conjunto de teste de acordo com o esperado
ind = np.where(y_test == 0.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color='violet', s=30, marker='^')

ind = np.where(y_test == 1.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color='lime', s=30, marker='X')

ind = np.where(y_test == 2.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color='brown', s=30, marker='d')

# Traçando a linha de regressão múltipla
x_values = np.linspace(np.min(X_train[:, 0]), np.max(X_train[:, 0]), 100)
y_values = np.linspace(np.min(X_train[:, 1]), np.max(X_train[:, 1]), 100)
X_mesh, Y_mesh = np.meshgrid(x_values, y_values)
Z_mesh = model.predict(np.column_stack((X_mesh.ravel(), Y_mesh.ravel())))
Z_mesh = Z_mesh.reshape(X_mesh.shape)

# Plotando a linha de regressão múltipla
plt.contour(X_mesh, Y_mesh, Z_mesh, colors='black', linewidths=0.5)

plt.show()

# Cálculo do score R2
score = r2_score(y_test, predictions)
print(score)

# Cálculo do RMSE
rmse = np.sqrt(((predictions - y_test) ** 2).mean())
print(rmse)
```
