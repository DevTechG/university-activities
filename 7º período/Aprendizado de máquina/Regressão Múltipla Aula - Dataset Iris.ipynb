```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression 


iris = load_iris()

#Armazenando os atributos na matriz X
xIris = iris.data
#Armazenando o vetor alvo em Y
yIris = iris.target
yIris = yIris.reshape(-1,1)
# Divisão dos dados
X_train,X_test,y_train,y_test = train_test_split(xIris[:,2:4],yIris, test_size = 0.20) 

degree = 2

model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X_train, y_train)
predictions = model.predict(X_test)

ind = np.where(y_train == 0)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind, 0], X_train[ind, 1], color = 'navy', s = 30, marker = 'o')

ind = np.where(y_train == 1)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind, 0], X_train[ind, 1], color = 'red', s = 30, marker = 'o')

ind = np.where(y_train == 2)[0]
ind = np.vstack(ind)
plt.scatter(X_train[ind, 0], X_train[ind, 1], color = 'green', s = 30, marker = 'o')

#Plotando o conjunto de teste de acordo com as previsões feitas
predictions = np.round(np.abs(predictions))
ind = np.where(predictions == 0.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color = 'purple', s = 30, marker = '*')

ind = np.where(predictions == 1.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color = 'purple', s = 30, marker = '^')

ind = np.where(predictions == 2.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color = 'purple', s = 30, marker = 'X')

#Plotando o conjunto de teste de acordo com o resultado esperado
ind = np.where(y_test == 0.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color = 'gray', s = 30, marker = '1')

ind = np.where(y_test == 1.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color = 'gray', s = 30, marker = '2')

ind = np.where(y_test == 2.0)[0]
ind = np.vstack(ind)
plt.scatter(X_test[ind, 0], X_test[ind, 1], color = 'gray', s = 30, marker = '3')

score = r2_score(y_test, predictions)
print(score)

rmse = np.sqrt(((predictions - y_test)**2).mean())
print(rmse)

plt.show()
```
